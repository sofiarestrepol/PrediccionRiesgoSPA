### Codificación One Hot


Se realiza una primera codificación para las columnas con multiples opciones de respuesta


```python
# Definir variables categóricas que serán codificadas usando el metodo One Hot Encoder
columnas_cat = ['Historial Familiar',
                'Condición',
                'Sustancias Vida', 
                'Sustancias Frecuente', 
                'Efectos Positivos Cannabis', 
                'Efectos Negativos Cannabis', 
                'Efectos Positivos Psilocibina', 
                'Efectos Negativos Psilocibina', 
                'Efectos Positivos MDMA', 
                'Efectos Negativos MDMA', 
                'Razón Tratamiento']
```

```python
# Se separan las diferentes opciones de respuesta por cada fila en cada una de las columnas categóricas
for col in columnas_cat:
    df_combined[col] = df_combined[col].str.split(';')

df_combined.head(2)
```

```python
# Crear columnas binarias usando 'explode' para descomponer las listas en filas
for col in columnas_cat:
    df_combined = df_combined.explode(col)

mostrar_df(df_combined)
```

```python
# Aplica One-Hot Encoding a las columnas categóricas del DF con los datos explotados
df_encoded = pd.get_dummies(df_combined, columns=columnas_cat, prefix=columnas_cat)
```

```python
# Agrupar por el índice original para reconstruir el DataFrame en la forma deseada
df_encoded = df_encoded.groupby(df_encoded.index).max().reset_index(drop=True)
mostrar_df(df_encoded)
```
